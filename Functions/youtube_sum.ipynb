```python
youtube_url = "https://www.youtube.com/watch?v=Y7W41VMxyQE&t=1s"
```

```python
from langchain_community.document_loaders import YoutubeLoader
from pytube import YouTube

content = ""
loader = YoutubeLoader.from_youtube_url(youtube_url, add_video_info=True)

try:
    youtube_data = loader.load()

    for doc in youtube_data:
        content += doc.page_content
        print(doc.page_content)

except Exception as e:
    print(f"Error loading video data: {e}")

    yt = YouTube(youtube_url)
```

```python

from langchain.vectorstores import FAISS
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings


text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
texts = text_splitter.split_documents(youtube_data)

embeddings = HuggingFaceEmbeddings(model_name="distilbert-base-uncased")

result_vectors = embeddings.embed_documents([t.page_content for t in texts])

db = FAISS.from_documents(texts,embeddings)
```

```python
content
```

```python
from langchain_google_genai import GoogleGenerativeAI

model = "gemini-little"
llm = GoogleGenerativeAI(model=model)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain.chains import create_retrieval_chain




prompts = ChatPromptTemplate.from_template(
    """
      Please summarize the content of the video based on the following context: {context}
    """
)

document_chain = create_stuff_documents_chain(llm=llm, prompt=prompts)
retriever = db.as_retriever()
retrieval_chain = create_retrieval_chain(retriever, document_chain)
```

```python
result = retrieval_chain.invoke({"input": content})


if 'text' in result:
    print(result['text'])
else:
    print("Result içinde 'text' anahtarı bulunamadı.")
```
