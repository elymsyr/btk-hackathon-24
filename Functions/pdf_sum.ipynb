```python
from IPython.display import Markdown,display
import textwrap

def to_markdown(text):
  text = text.replace("*", "\n")
  return Markdown(textwrap.indent(text, ">  ",predicate=lambda _:True))
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader('gezegenler.pdf')

pdf_data = loader.load()



from langchain_core.prompts import ChatPromptTemplate
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_google_genai import ChatGoogleGenerativeAI


prompt = ChatPromptTemplate.from_template("""
    You are an AI assistant that specializes in summarizing documents.
    Your task is to read the given context and produce a concise and informative summary.
    Make sure the summary captures the key points and avoids unnecessary details.
    Here's the context: {context}
""")

model = 'gemini-pro'


llm = ChatGoogleGenerativeAI(
    model=model,
    temperature=0.7,
    max_output_tokens=1024,
    top_p=0.9,
    top_k=40,
    repetition_penalty=1.2)


document_chain = create_stuff_documents_chain(llm, prompt)

summary = document_chain.invoke({"context": pdf_data})


print(summary)
```
